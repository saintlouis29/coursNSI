# Utiliser un python virtuel

La commande *python* appelle l'interpréteur python installé sur le système.

Vous pouvez lancer la commande suivante pour vous en convaincre :

```
which python
```

Lorsque vous developpez un programme vous avez souvent besoin de librairies externes (Image, son, IA...) qui ne sont pas forcément installées sur le système.

- Si vous n'êtes pas administrateur système, nous ne pouvez pas installer ces librairies.
- Si vous êtes admnistrateur système, vous n'avez pas envie d'installer des librairies qui peuvent avoir des conséquences sur l'ensemble du système...

La librairie *python venv* permet de créer un environnement python dans lequel vous pouvez ajouter vos propres librairies sans avoir besoin de les installer sur le système.

## Quant utiliser un environnement virtuel ?

Toujours ! Sauf si pour un petit projet vous n'avez pas besoin de librairie externe.

## Comment installer et configurer son environnement virtuel ?

<div class="alert alert-info">Nous allons prendre l'habitude d'installer nos environnements virtuels toujours dans le même dossiers, à savoir le dossier <b>~/venv</b> .</div>

### Installation

```
python3 -m venv ~/venv/nom_de_votre_environnement
```
<div class="alert alert-info">le symbole <b>~</b> est un raccourci pour désigner <b>votre répertoire home</b>.</div>

Cette commande créée un environnement python complet issu du python système, vous avez la même version, mais les librairies sont installées dans le dossier du python virtuel.

### Activation de l'environnement
```
source ~/venv/nom_de_votre_environnement/bin/activate
```

Votre terminale vous indique alors quel python virtuel vous utilisez.

```
(testpy) y.cadour@workstation0 ~ $ 
```

<div class="alert alert-info">Si vous utilisez un IDE, il est nécessaire de lui indiquer quel python vous utilisez</div>

Une fois cet environnement installé, vous pouvez installer les librairies dont vous avez besoin pour votre projet.

```
pip install Sphinx
```

**Remarque**

pip (Python Installer Package ) est le gestionnaire d'installation des paquets python.

Pour plus d'infos, aller sur le [site](https://pip.pypa.io/en/stable/) ou utiliser la commande suivante :

```
man pip
```


### désactivation de l'environnement

Il suffit de lancer la commande suivante :
```
deactivate
```
La commande *python* est désormais celle du système.

### suppression de l'environnement

La commande suivante supprime votre environnement:
```
rm ~/venv/nom_de_votre_environnement -rf
```
