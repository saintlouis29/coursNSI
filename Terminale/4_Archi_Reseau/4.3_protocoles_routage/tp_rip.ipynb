# TP RIP

*Ce TP est en relation avec le [cours sur les protocoles réseau](protocoles_routage.ipynb).*

Pour étudier le protocole RIP, nous allons utiliser le logiciel open source [GNS3](https://www.gns3.com/).

Le matériel que nous allons utiliser comporte:

- des routeurs Cisco;
- des PC virtuels basiques;
- différents types de liaisons (fast ethernet, ethernet, gigabits ethernet...)

- **Lancer le logiciel GNS3**

## Configuration du routeur R6

### Démarrage des interfaces:

Voici comment on été démarrées les interfaces de R1, **adaptez** les au routeur R6.

```
R1#
R1#configure term
Enter configuration commands, one per line.  End with CNTL/Z.
R1(config)#interface g1/0
R1(config-if)#ip address 10.1.1.254 255.255.255.0
R1(config-if)#no shutdown
R1(config-if)#interface g2/0
R1(config-if)#ip address 1.1.1.1 255.255.255.0
R1(config-if)#no shutdown   
R1(config-if)#interface f0/0
R1(config-if)#ip address 6.1.1.1 255.255.255.0
R1(config-if)#no shutdown   
R1(config-if)#exit
```
Nous pouvons maintenant vérifier si les interfaces sont correctement démarrées.

```
R6#show ip interface brief
```

### configuration du protocole RIP

Voici comment a été configuré R1, **adaptez** cela pour le routeur R6.

```
R1#config term
Enter configuration commands, one per line.  End with CNTL/Z.
R1(config)#router rip
R1(config)#version 2
R1(config-router)#network 10.1.1.0  
R1(config-router)#network 1.1.1.0   
R1(config-router)#network 6.1.1.0   
R1(config-router)#exit
```

### sauvegarde de la config

Pour sauvegarder la configuration du routeur, tapez la commande ci-dessous.


```
copy running-config startup-config

```
