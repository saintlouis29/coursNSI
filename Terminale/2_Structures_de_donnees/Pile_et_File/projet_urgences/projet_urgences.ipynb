# Projet Urgences

**Ce projet est fortement inspiré de celui d'un collègue de NSI, Jocelin Devalette, professeur de Physique/NSI au lycée français de Tananarive.**

## Contexte

Vous travaillez dans un service d'urgences hospitalier où les patients sont triés selon leur niveau de gravité, de 0 à 4. Le niveau 0 est le plus urgent, le 4 le plus léger. Les urgentistes traitent les patients un par un, en commençant par les plus urgents.

- Les patients arrivent à différents moments avec des niveaux de gravité variables
- Le service doit compter les patients par gravité pour ajuster les ressources

Vous devez modéliser la structure adéquate afin que les deux règles suivantes soient respectées.

- Priorité par gravité : les patients sont prioritairement traités par gravité. (gravité 0 en premier)
- Ordre d'arrivée : Pour les patients de gravité égale, on suit l'ordre FIFO.

## Modélisation

Créez un projet *Urgences* dans lequel vous implémenterez trois classes

 - *StructureUrgences*. (L'utilisation du module sl29.structures.queue est fortement encouragée...)
 - *Patient*. Afin de découvrir une nouvelle technique de codage des classes, jetez un oeil sur les [property en Python](https://www.docstring.fr/glossaire/propriete/) (hors programme)
 - *GestionUrgences*

Voici les diagrammes des classes.

Une fois ceci implémenté, vous devez ajouter une méthode *simuler_urgences* dont le but est d'estimer le nombre d'urgentistes nécessaires.

Voici les conditions à respecter:

- Le temps de traitement moyen d’un patient est de :
  - 75 minutes pour un patient de gravité 0
  - 60 minutes pour un patient de gravité 1
  - 45 minutes pour un patient de gravité 2
  - 30 minutes pour un patient de gravité 3
  - 15 minutes pour un patient de gravité 4.
- Un patient arrive en moyenne toutes 10 minutes au service des urgences.
- La file d’attente des urgences ne peut accueillir que 50 patients au maximum, si ce nombre est atteint il faut affecter un nouveau médecin.
- Un patient de gravité 0 ou 1 ne peut attendre, si aucun médecin n'est disponible, il faut en affecter un nouveau.
- Une [liste de 200 patients aléatoire est mise à disposition](liste_patients.py), cette liste doit être mélangée avec la fonction
shuffle au début de chaque exécution de la simulation.
- Le nombre de médecins dans le service est initialisé à 1, il est possible de changer cette valeur ! Les
médecins sont représentés par un dictionnaire dont la clé est le nom du médecin (médecin1, médecin2...) et la valeur un tuple (patient, temps de traitement restant) ou (None, 0) si
aucun patient ne lui est affecté.
- la simulation affiche chaque minute l'ensemble des médecins et leur patient ainsi que la durée restante de traitement
- La simulation simule une garde de 12H (720 minutes).

Vous utilisez un compteur qui représente les minutes de la garde. Vous pourrez utiliser le modulo pour ajouter un nouveau patient dans la structure toutes les 10 minutes.
