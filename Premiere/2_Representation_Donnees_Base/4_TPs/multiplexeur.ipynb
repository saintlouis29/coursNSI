# TP Multiplexeur

Un **multiplexeur** (abréviation : MUX) est un circuit permettant de concentrer sur une même voie de transmission différents types de liaisons (informatique, télécopie, téléphonie, télétex...) en sélectionnant une entrée parmi N.

Il posséde donc **une sortie** et **N entrées**, ainsi que **log2 N entrées de commande** permettant de choisir quelle entrée sera sélectionnée.

Nous allons étudier un multiplexeur 4:1
Les  entrées sont  X0, X1, X2, X3..

Combien d’entrées de commande sont nécessaires pour choisir une entrée parmi les 4?

Et si il y avait 16 entrées?

On appelle A et B les deux commandes et Z la sortie.
